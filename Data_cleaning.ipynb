Data overview

```python
import pandas as pd
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load and explore the spam data

```python
data = pd.read_csv('/content/messages.csv',sep='\t',names=['label','message'])
data.head(5)
```

```python
data['label'].value_counts()
```

Clean the Data (punctuation removal)

```python
import string
string.punctuation
def remove_punctuation(text):
    punctuation_free="".join([i for i in text if i not in string.punctuation])
    return punctuation_free
data['Cleaned_message'] = data['message'].apply(lambda x: remove_punctuation(x))
```

```python
data.head(10)
```

Tokenization

```python
import re

def tokenizer(text):
  tokens = re.split('\W+',text)
  return tokens

data['Tokenized_data'] = data['Cleaned_message'].apply(lambda x: tokenizer(x))
```

```python
data.head(5)
```

Stop words removal

```python
import nltk
nltk.download('stopwords')
st_words = nltk.corpus.stopwords.words('english')

print(st_words[:30])
def rem_st(text):
  return [i for i in text if i not in st_words]

data['Removed_st_word'] = data['Tokenized_data'].apply(lambda x: rem_st(x))
```

```python
data.head(5)
```

**Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()

def lem_text(text):
  return [wordnet_lemmatizer.lemmatize(word) for word in text]

data['Lemitized_text']  = data['Removed_st_word'].apply(lambda x: lem_text(x))

data.head(5)
```

Vectorizer

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()
doc_vec = tfidf_vectorizer.fit_transform(data['Lemitized_text'].iloc[1])
```

```python
print(doc_vec)
```

```python
vec_data = pd.DataFrame(doc_vec.toarray().transpose(),
                   index=tfidf_vectorizer.get_feature_names_out())
```
